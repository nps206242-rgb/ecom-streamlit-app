####In Data Science, data type must match the type of data.
If the data type is wrong, analysis, calculations, and graphs may not work properly.

Example:

**If numbers are stored as text (object), Python cannot perform math operations.**



Upon examining the dataset, several columns have incorrect data types that need to be updated to ensure proper analysis and visualization. Here are the necessary adjustments:  

1. **Reviews**:  
   - **Current Data Type**: Object  
   - **Required Data Type**: Integer or Float  
   - **Reason**: The number of reviews should be a numerical value for aggregation and statistical analysis.  

2. **Size**:  
   - **Current Data Type**: Object  
   - **Required Data Type**: Float or Integer  
   - **Reason**: The app size, often in KB or MB, is numerical data and needs to be converted to facilitate comparisons and calculations.  

3. **Installs**:  
   - **Current Data Type**: Object  
   - **Required Data Type**: Integer  
   - **Reason**: Install count is a numeric value and should be converted to enable meaningful sorting, filtering, and analysis.  

4. **Price**:  
   - **Current Data Type**: Object  
   - **Required Data Type**: Float  
   - **Reason**: Prices, even in string format (e.g., "$4.99"), need to be converted into numerical values for computations. The dollar symbol can be stripped during preprocessing.  

5. **Last Updated**:  
   - **Current Data Type**: Object  
   - **Required Data Type**: DateTime  
   - **Reason**: Dates are crucial for time-based analysis and trends. Converting this to a DateTime format ensures proper sorting and time-related calculations.  



These changes will enhance the dataset's usability and ensure more accurate and meaningful analyses.

````python
# Converting Reviews column dataype to numeric

def convert_reviews(reviews):
    # Checking if the value contains 'M' (indicating millions)
    if 'M' in reviews:
        # Remove 'M', convert the remaining part to float and multiply by 1,000,000(1m)
        return float(reviews.replace('M', '')) * 1000000

    # Checking if the value contains 'K' (indicating thousands)
    elif 'K' in reviews:
        # remove 'K', convert the remaining part to float and multiply by 1,000(1k)
        return float(reviews.replace('K', '')) * 1000

    # If the value does not contain 'M' or 'K', assuming it is already numeric
    else:
        return float(reviews)

print('Before coverting data type of reviews column:',df['Reviews'].dtypes)
df['Reviews'] = df['Reviews'].apply(convert_reviews) # using apply() to apply the convert_reviews() function to all values of Reviews column
df['Reviews'] = df['Reviews'].astype(int) # in reviews column float is not a good option, so converting it to int
print('After coverting data type of reviews column:',df['Reviews'].dtypes)
````

````python
# Converting Size column datatype to numeric in MB

def convert_size(size):
    # Checking if the size is in megabytes (indicated by 'm' or 'M')
    if 'm' in size.lower():
        # Retain the value in MB directly
        return float(size.lower().replace('m', '').replace(',', ''))
    # Checking if the size is in kilobytes (indicated by 'k' or 'K')
    elif 'k' in size.lower():
        # Convert the kilobytes to megabytes
        return float(size.lower().replace('k', '').replace(',', '')) / 1024
    # Checking if the size contains a '+' (e.g., '1,000+')
    elif '+' in size:
        # Convert to float and treat it as MB (you can adjust if needed)
        return float(size.replace(',', '').replace('+', ''))
    # Handle entries with 'Varies with device', replacing them with NaN
    elif size.lower() == 'varies with device':
        return np.nan
    # Handling any unexpected formats by returning NaN
    else:
        return np.nan

# Apply the function to the 'Size' column
print('Before converting datatype of Size column:', df['Size'].dtypes)
df['Size'] = df['Size'].apply(convert_size)
df.rename({'Size': 'Size (MB)'}, axis=1, inplace=True)
print('After converting datatype of Size column:', df['Size (MB)'].dtypes)
````

````python
# Converting Installs column dataype to numeric

def convert_installs(installs):
    # Handle 'Free' case and replace it with 0
    if installs.lower() == 'free':
        return 0
    # Remove the comma and the '+' symbol, convert to integer
    elif '+' in installs:
        return int(installs.replace(',', '').replace('+', ''))
    # Handling '0+' or '0' case as 0 installs
    elif installs == '0+' or installs == '0':
        return 0
    # Otherwise, removing the comma and return as integer
    else:
        return int(installs.replace(',', ''))

print('Before coverting data type of installs column:',df['Installs'].dtypes)
df['Installs'] = df['Installs'].apply(convert_installs) # Appling the function to the 'Installs' column
print('After coverting data type of installs column:',df['Installs'].dtypes)
````

````python
# Converting Price column dataype to numeric

def process_price(price):
    # Handle '0' (Free apps)
    if price == '0':
        return 0.0
    # Removing the dollar sign ('$') and convert to float
    elif price.startswith('$'):
        return float(price.replace('$', ''))
    # Handling invalid entries like 'Everyone'
    else:
        return np.nan

# Applying the function to the 'Price' column
print('Before coverting data type of price column:',df['Price'].dtypes)
df['Price'] = df['Price'].apply(process_price)
df.rename({'Price':'Price ($)'},axis=1 , inplace=True)
print('After coverting data type of price column:',df['Price ($)'].dtypes)
````

````python
# Convert 'Last Updated' column to datetime, coercing errors
print('Before coverting data type of Last Updated column:',df['Last Updated'].dtypes)
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')

print('After coverting data type of Last Updated column:',df['Last Updated'].dtypes)
````
